```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("/kaggle/input/symptoms-and-covid-presence/Covid Dataset.csv")
df.head()
```

```python
df_modified = df.copy()
# making special labelencoder instance for target variable to have consistent
# translation between text -> int encoding/decoding for all datasets
target_encoder = LabelEncoder().fit(df_modified['COVID-19'])
df_modified['COVID-19'] = target_encoder.transform(df_modified['COVID-19'])
# converting from categorical data to numerical
for column in df_modified.columns.astype('category'):
 df_modified[column] = df_modified[column].astype('category').cat.codes
df_modified.head()
```

```python
x = df_modified.drop('COVID-19', axis=1)
y = df_modified['COVID-19']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
print('Train shape', x_train.shape, y_train.shape)
print('Train shape', x_test.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
tr_predict = model.predict(x_train)
ts_predict = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print('Training accuracy : ', accuracy_score(y_train, tr_predict))
print('Testing accuracy : ', accuracy_score(y_test, ts_predict))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize = (25, 20))
plot_tree(model, filled = True, feature_names = x_train.columns, max_depth = 5, fontsize = 12)
plt.show()
```

```python
df_importance = pd.DataFrame({'Features': x.columns, 'Importances' : model.feature_importances_})
df_importance
```
